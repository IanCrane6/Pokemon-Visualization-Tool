```python
from ultralytics import YOLO
import torch
```

```python
model = YOLO("yolo11s.pt")

model = model.train(data = 'PokeYOLO.yaml', epochs = 100, imgsz = (224, 224), batch = 16,
                    name = "PokeSnapper", project = 'runs/detect', patience = 20, device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu"))
```

```python
model = YOLO('/Users/iancrane/Desktop/CS-7643/Project/Pokemon-Visualization-Tool/runs/detect/runs/detect/PokeSnapper2/weights/best.pt')

results = model.predict('/Users/iancrane/Desktop/CS-7643/Project/Pokemon-Visualization-Tool/Dataset/PokemonAPI/Pikachu/back_default.png', conf = 0.25)

for r in results:
    r.show()
    r.save("output.jpg")
    
for r in results:
    boxes = r.boxes.xyxy
    confs = r.boxes.conf
    classes = r.boxes.cls
```
